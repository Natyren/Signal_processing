```python
import numpy as np
import matplotlib.pyplot as plt

import sounddevice as sd
```

```python
class CFG:
    f = 2000


freq = {'0': [1336, 941],
        '1': [1209, 697], '2':[1336, 697], '3': [1477, 697],
        '4': [1209, 770], '5':[1336, 770], '6': [1477, 770],
        '7': [1209, 852], '8':[1336, 852], '9': [1477, 852]}

def dtfm(n):
    fhigh, flow = freq[n]
    t = np.linspace(0, 0.500, int(0.5*CFG.f))
    return np.sin(2*np.pi*fhigh*t)+np.sin(2*np.pi*flow*t)

def playseq(s):
    return np.array([dtfm(n) for n in s]).ravel()

def plot():
    plt.figure(figsize=(20, 3))
    plt.plot(playseq(seq))
    plt.show()
```

```python
seq = list(input('Введите последовательность чисел '))
```

```python
plot()
```

```python
sd.play(playseq(seq), CFG.f)
```
